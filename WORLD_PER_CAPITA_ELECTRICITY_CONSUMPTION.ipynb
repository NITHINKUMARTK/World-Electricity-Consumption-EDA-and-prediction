# **TRAINING THE MODEL**

***TRAINING THE MODEL FOR PREDICTING INDIA'S CONSUMPTION***

```python
india_consumption
```

```python
X_india = india_consumption.iloc[:,1].values.reshape(36,1)
y_india = india_consumption.iloc[:,-1].values

```

```python
print(X_india.shape)
print(X_india)
```

```python
print(y_india.shape)
print(y_india)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_india,y_india,test_size = 1/3, random_state = 0)
print(X_train)
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model_india = LinearRegression()
model_india.fit(X_train,y_train)
```

```python
print("Coefficient: ",model_india.coef_)
print("intercept: ",model_india.intercept_)
```

```python
y_pred = model_india.predict(X_test)
```

***VISUALIZING TRAINING SET RESULTS***

```python
plt.scatter(X_train, y_train, color = 'green',label = "Actual value")
plt.plot(X_train, model_india.predict(X_train), color = 'red', label = "Predicted value")
plt.title('INDIA(Training data)',color= "blue")
plt.xlabel('Year',color = "red")
plt.ylabel("per capital electricity consumption(kWh)",color = "red")
plt.legend(loc= 0)
plt.show()
```

***VISUALIZING THE TEST RESULTS***

```python
plt.scatter(X_test, y_test, color = 'green',label = "Actual value")
plt.plot(X_train, model_india.predict(X_train), color = 'red', label = "Predicted value")
plt.title('INDIA(Testing data)',color= "blue")
plt.xlabel('Year',color = "red")
plt.ylabel("per capital electricity consumption(kWh)",color = "red")
plt.legend(loc = 0)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
error = mean_squared_error(y_test,y_pred, squared = True)
print(error)
```

```python
years=[2021,2022,2023,2024,2025,2036,2037,2040]
for i in years:
  print(model_india.predict([[i]]))
```
